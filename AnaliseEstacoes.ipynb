```python
import pandas as pd
import os
```

```python
df_estacoes_automaticas = pd.read_csv("dados_inmet/CatalogoEstaçõesAutomáticas.csv", sep=";", decimal=",")
df_estacoes_convencionais = pd.read_csv("dados_inmet/CatalogoEstaçõesConvencionais.csv", sep=";", decimal=",")
```

```python
df_estacoes_automaticas_mg = df_estacoes_automaticas.loc[
    (df_estacoes_automaticas["SG_ESTADO"] == "MG") &
    (df_estacoes_automaticas["CD_SITUACAO"] == "Operante")
]
df_estacoes_automaticas_mg.reset_index(drop=True, inplace=True)
df_estacoes_automaticas_mg['DT_INICIO_OPERACAO'] = pd.to_datetime(df_estacoes_automaticas_mg['DT_INICIO_OPERACAO'], format="%d/%m/%Y")

df_estacoes_convencionais_mg = df_estacoes_convencionais.loc[
    (df_estacoes_convencionais["SG_ESTADO"] == "MG") &
    (df_estacoes_convencionais["CD_SITUACAO"] == "Operante")
]
df_estacoes_convencionais_mg.reset_index(drop=True, inplace=True)
df_estacoes_convencionais_mg['DT_INICIO_OPERACAO'] = pd.to_datetime(df_estacoes_convencionais_mg['DT_INICIO_OPERACAO'], format="%d/%m/%Y")
```

```python
codigos_estacoes = df_estacoes_automaticas_mg["CD_ESTACAO"].to_list()
anos = [2019, 2020, 2021, 2022, 2023, 2024]

for codigo in codigos_estacoes:
    csv_files = []
    for ano in anos:
        files = os.listdir(f"dados_inmet/{ano}")
        for file in files:
            if f"MG_{codigo}" in file:
                df_temp = pd.read_csv(f"dados_inmet/{ano}/{file}", sep=";", decimal=",", encoding="latin-1", skiprows=8)
                df_temp.drop("Unnamed: 19", axis='columns', inplace=True)
                df_temp.columns = df_temp.columns.str.upper()
                df_temp.fillna(0, inplace=True)
                
                csv_files.append(df_temp)
                break
    df_result = pd.concat(csv_files)
    df_result.to_csv(f"dados_inmet/Dados meteorológicos Estação {codigo}-{df_estacoes_automaticas_mg.loc[df_estacoes_automaticas_mg["CD_ESTACAO"] == codigo, "DC_NOME"].iloc[0]}.csv", index=False, sep=";", decimal=",")
```
